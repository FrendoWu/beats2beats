```python
import midi
import operator
import numpy as np
import matplotlib.pyplot as plt
from track_util import *
from util import *
```

```python
def count_other_event(track):
    count = 0
    for event in track:
        if not isinstance(event, midi.NoteOnEvent):
            count = count + 1
        else:
            return count
```

```python
midiname = "../track/The_Real_Slim_Shady_0_Harpsichord.mid"
pattern = read_midifile(midiname)
matrix = track_to_matrix(pattern[1])
matrix = sort_matrix(matrix)
```

```python
start = count_other_event(pattern[1])
```

```python
t = matrix[:,0]
t = t[1:] - t[:-1]
plt.plot(range(len(t))[:120],t[:120])
```

```python
m = max(t) / 3
```

```python
peak = []
for idx, value in enumerate(t):
    if value > m:
        peak.append(idx)
```

```python
t = matrix[:,0]
delta = []
for i in range(len(peak)-1):
    delta.append(t[peak[i+1]]-t[peak[i]])
```

```python
plt.plot(range(len(delta))[:10],delta[:10])
```
